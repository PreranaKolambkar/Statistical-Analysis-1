# Conversion of Categorical Data into Dummy Variables

```python
catg_col= [i for i in df.columns if df[i].dtype =='O']
```

```python
catg_col
```

#ONE HOT ENCODER

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one_hot=OneHotEncoder()
```

```python
one_hot.fit_transform(df[catg_col]).toarray()
```

```python
df2=pd.DataFrame(one_hot.fit_transform(df[catg_col]).toarray())
```

```python
df2
```

```python
df2.columns = one_hot.get_feature_names_out(catg_col)   # to get meaningful column name
```

```python
df3=df1.join(df2)
```

```python
df3
```

```python
df3.drop(columns=['SKU', 'City', 'BU', 'Brand', 'Model'],inplace=True)
```

# Conclusion

Standardization ensures that all numerical features are on the same scale, which helps ML algorithms to perform better

One-hot encoding converts categorical variables into binary (0 or 1) columns, which makes them easier for machine learning models to process.
